```python
corpus = []

for i in range(0,len(data)):
    corpus.append(data[i]["Text"])

corpus[:10] # list of all documents
```

```python
corpus[0].lower().split(" ")
```

```python
stoplist = set('for a of the and to in'.split(' '))

texts = [[word for word in document.lower().split(" ") if word not in stoplist] for document in corpus]
```

```python
texts[0:2]
```

```python
from collections import defaultdict
frequency = defaultdict(int)
print(frequency)
```

```python
for text in texts:
    for token in text:
        frequency[token] += 1

for i,j in frequency.items():
    print(i,j)
```

```python
processed_output = [[word for word in text if frequency[token] > 1] for text in texts]
```

```python
frequency_count_1_words = [token for token , count in frequency.items() if count == 1]
```

```python
frequency_count_1_words # there were so such tokens which have frequency as 1 
```

```python
processed_output[:5]
```

```python
from gensim import corpora
dictionary = corpora.Dictionary(processed_output)
```

```python
dictionary.save('/tmp/deerwester.dict') 
```

```python
print(dictionary)
```

```python
print(len(dictionary))
```

```python
count = 0
for token , id in dictionary.token2id.items():
    count += 1
    print(token , id)
    if count == 10:
     break
```

```python
print(dictionary.token2id)
```

```python
bow_corpus = [dictionary.doc2bow(text) for text in processed_output]
```

```python
bow_corpus[:5] # token id in vocab , frequency count of word in document
```

```python
from gensim import models

tfidf = models.TfidfModel(bow_corpus) # based on the rareity of a token 

```

```python
print(tfidf[dictionary.doc2bow("He is so mean to me , I can't say he he he ".lower().split(" "))])
```

```python
print([dictionary.doc2bow("He is so mean to me , I can't say he he he ".lower().split(" "))])
```

```python
tf_idf_values = [tfidf[dictionary.doc2bow(text)] for text in processed_output]
```

```python
tf_idf_values[:2] # token id , tf-idf value
```

```python
from gensim import similarities

index = similarities.SparseMatrixSimilarity(tf_idf_values, num_features=49134)
```

```python
query_document = 'I really love you friends , we will be there for each other forever'.split()
query_bow = dictionary.doc2bow(query_document)
sims = index[tfidf[query_bow]]
print(list(enumerate(sims)))
```

```python
count = 0
for document_number, score in sorted(enumerate(sims), key=lambda x: x[1], reverse=True):
    count +=1
    print(f"{document_number, score , processed_output[document_number]}")
    if count ==20:
        break
# bow and tf-idf methods are purely driven by frequencies in the document and similarties are also found in terms of frequency of words present in query against present in the documents in the database.    
```

```python
# The core concepts of gensim are:

# Document: some text.

# Corpus: a collection of documents.

# Vector: a mathematically convenient representation of a document.

# Model: an algorithm for transforming vectors from one representation to another.
```

```python
# Using streaming that is not to load all of the data directly into the RAM and instead accessing one at a time -> streaming (just have to define the iter function correctly as per our application)
# The corpus is now much more memory friendly, because at most one vector resides in RAM at a time. Your corpus can now be as large as you want. -> Using streaming
```

```python
# Corpus objects are streams, so typically you won’t be able to print them directly
```

```python
# another way of doing it: print one document at a time, making use of the streaming interface
# for doc in corpus:
#     print(doc)
# The second way is obviously more memory-friendly, but for testing and development purposes, nothing beats the simplicity of calling list(corpus).
# In this way, gensim can also be used as a memory-efficient I/O format conversion tool: just load a document stream using one format and immediately save it in another format.
```
